```python
import matplotlib.pyplot as plt

cols_to_plot = df.select_dtypes(include="number").columns

# Plot histograms
df[cols_to_plot].hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
df["datum"] = pd.to_datetime(df["datum"])

for col in df.columns:
    if col == "datum":
        continue

    plt.figure(figsize=(10, 4))
    plt.plot(df["datum"], df[col])
    plt.title(col)
    plt.xlabel("Date")
    plt.ylabel(col)
    plt.grid(True)
    plt.show()
```
